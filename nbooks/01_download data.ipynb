### Connection to Quandl API

We want to download sample data from Quandl API, which provides sample datasets for some of the S&P500 indices. Â£ functions are built in order to:
* build a url for a get request bound ti a specific index
* process a get request and save the content to a json file, if the dataset is available. Otherwise, store the non-available ticker into a variable
* run the process for all the S&P500 costituents

NOTE: We omit dates, we will process these later

```python
global non_av_tickers
non_av_tickers = []

def quandl_url(ticker):
    """ Build a url for a get request to quandl, related to a ticker
    """
    url = r'https://www.quandl.com/api/v3/datasets/EOD/'
    params={"api_key":"jnwio6ujAZibr4cC7Xts"}
    params = urlencode(params, quote_via=quote_plus)
    
    return url+ticker+"?"+params

def quandl_get(url, fn):
    """ Save a request payload as json
    """
    r = requests.get(url)
    if r.text != '{"quandl_error":{"code":"QEPx04","message":"You do not have permission to view this dataset. Please subscribe to this database to get access."}}':
        with open(fn,'w') as f:
            json.dump(json.loads(r.text), f)
    else:
        t = fn.split('\\')[-1]
        print('{} could not be saved'.format(t))
        non_av_tickers.append(t.split('.')[0])

def download_snp(df):
    """ Downalod all s&p500 as json files, when available
    """
    tickers = df["Symbol"].unique()
    for t in tickers:
        quandl_get(quandl_url(t), os.path.join(data_folder,t+'.json'))
```
